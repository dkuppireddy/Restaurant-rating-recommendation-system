```python
#Importing required Libraries
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')

from nltk.corpus import stopwords
```

```python

#--------Loading dataset---------


restaurant_data=pd.read_csv("/Users/kumar/Documents/ADT/google.csv")
data_df=restaurant_data.copy()
data_df
```

```python
#------Remove the NaN values from the dataset-----------
data_df.dropna(how='any',inplace=True)
data_df
```

```python
#----------CHANGING COLUMN NAMES-----------
data_df=data_df.rename(columns={'vzRMeAg4Oi8__kl-trunc-name':'Restaurant_Name','sBhnyP5sXkG__rtng':'Ratings',
                                'sBhnyP5sXkG__number-of-reviews':'review_count','tLipRb src':'url',
                                'rllt__details':'cuisine','rllt__details 3':'address',
                                'BI0Dve':'Type1','BI0Dve 2':'Type2','BI0Dve 3':'Type3','rllt__wrapped':'reviews'})
print('Column names Changed')
```

```python
data_df
```

```python
#---To merge Type1,Type2,Type3 columns, we need to remove columns between them----
data_df=data_df.drop(['if66xd','if66xd 2'],axis=1)
data_df
```

```python
#-----Merging 3 columns into single column-------
data_df['Type'] = data_df[data_df.columns[11:13]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
data_df
```

```python
#----------DELETING UNWANTED COLUMNS---------
data_df=data_df.drop(['url','dbg0pd','YDIN4c','HypWnf','rllt__details 2','Type1','Type2','Type3'],axis=1)
data_df
```

```python
#---------AFTER CLEANING DATA SHAPE HAS BEEN CHANGED------
#NO null records
data_df.shape
```

```python
#-----------DATA TRANSFORMATION----------
```

```python
#-------Making list with all restaurant names in it-------
restaurants=list(data_df['Restaurant_Name'].unique())
restaurants
```

```python
#----Insert a new column called 'Mean Rating' in data frame for all records which is equal to 0---
data_df['Mean Rating']=0
data_df
```

```python
# CALCULATE MEAN RATING
for i in range(len(restaurants)):
    data_df['Mean Rating'][data_df['Restaurant_Name'] == restaurants[i]] = data_df['Ratings'][data_df['Restaurant_Name'] == restaurants[i]].mean()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (1,5))
data_df[['Mean Rating']] = scaler.fit_transform(data_df[['Mean Rating']]).round(2)
```

```python
data_df[['Restaurant_Name','Ratings','Mean Rating']].head()
```

```python
#----- TEXT PRE-PROCESSING AND CLEANING
```

```python
#----Lower Casing text in 'Reviews' coulmn----------
data_df["reviews"] = data_df["reviews"].str.lower()
data_df
```

```python
#------Removing Punctuations in 'Reviews' Column------
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

data_df["reviews"] = data_df["reviews"].apply(lambda text: remove_punctuation(text))

data_df
```

```python
#-------Removing Stop Words in 'Reviews' Column------
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

data_df["reviews"] = data_df["reviews"].apply(lambda text: remove_stopwords(text))

data_df
```
